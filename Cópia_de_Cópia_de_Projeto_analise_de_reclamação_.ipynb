  através dos consumidores coletar informações para diagnosticar o perfil das empresas no segmento do comércio eletrônico, e para obter informações foi pesquisado a base de dados do portal"Consumidor. gov" no período de julho de 2021. Tais informações buscam entender tanto os clientes e as empresas prestadoras dos seus  serviços. Entender os consumidores atrelado aos comportamento do processos de negócio busca uma melhor tomada das decisões por parte dos seus gestores, e consequentemente ganhos de competitividade.

```python
import pandas as pd  # importando bibliotecas
```

```python
dados = pd.read_csv('/content/2021-07.csv', sep=';') # abrindo dataset
display(dados)
```

```python
dados.shape # verificando a quantidadde de linha s e colunas
```

```python
dados.info() # Buscando informações sobre o dataset
```

```python
dados.isnull().sum().sort_values(ascending=False) # verificando a soma dos dados nulos
```

```python
média = dados['Nota do Consumidor'].mean() # tirando a média da nota do consumidor 
```

```python
dados.fillna(média, inplace=True )
```

```python
dados
dados.drop(12500) # foi dropada a linha que não continha dados suficientes!
```

```python
média2 = dados['Tempo Resposta'].median() # Calculando a médiana o tempo de respostas
```

```python
dados.fillna(média2, inplace=True ) # substituindo dados faltantes pela mediana 
dados
dados.drop(12500) 
dados.drop(91960)
```

```python
dados['Sexo']!=2.9960778643338695
novo_df=dados[dados['Sexo']!=2.9960778643338695]
novo_df
```

```python
novo_df['Sexo'].value_counts()  # valores masculino e feminino
```

```python
dados.fillna(novo_df, inplace=True) # Subtituindo valores faltantes
dados
dados.drop(12500) 
```

```python
dados.isnull().sum().sort_values(ascending=False) # verificando todos os dados faltantes
```

```python
ce=dados[dados['Segmento de Mercado']=='Comércio Eletrônico'] # Segmentando o mercado por comércio eletronico(ce)
ce
```

##Qual a média avaliada pelo consumidores por empresa no segmento comércio eletrônico?##


```python
ce.groupby('Nome Fantasia').mean()['Nota do Consumidor'].round(2) # Agrupando a nota do consumidor e tirando a média
```

```python
ce.groupby('Nome Fantasia').mean()['Nota do Consumidor'].round(2). plot(kind = 'barh', figsize=(14, 10),title='Média da nota do  consumidor');
# Agrupando e verificando a nota média do consumidor
```

## Qual as empresas com maiores números de reclamações no comércio eletrônicos?##


```python
ce['Nome Fantasia'].value_counts() # Fazendo a contagem das reclamações!
```

## Qual a média do tempo de resposta ao consumidor?##

```python
ce.groupby('Nome Fantasia').mean() ['Tempo Resposta'].sort_values(ascending = False) # Agrupando e tirando a média do tempo de respostas
```

## Qual a predominância dos gêneros dos consumidores?##

## Qual a faixa etária dos consumidores reclamante?##
